```python
# Read in the database
# DATABASE_FILE = 'database.csv'
DATABASE_FIELDS = ['year', 'tif_number', 'page_num', 'block_num', 'par_num', 'line_num', 'word_num', 'left', 'top', 'width', 'height', 'conf', 'text']
DATABASE_URL = 'https://noahbaxley.com/tiftastic/uploads/2010_13.csv'   # This is a url to one of the pre-processed documents
DATABASE_FILE = 'parsed_pdfs/2012_10.csv'
database = pd.read_csv(DATABASE_URL,header=None, names=DATABASE_FIELDS)
```

```python
# Sort all words so that they appear in order
database.sort_values(['page_num', 'block_num', 'line_num', 'word_num'])
database.head(n=15)
```

```python
# Concatenate all words
# By spaces in word_num, by carriage returns in line_num
doc_string = ''
current_page_number = 0
current_line_number = 1
current_blok_number = 1
for index, row in database.iterrows():
	if row['page_num'] > current_page_number:
		doc_string += '\n'
		doc_string += f'\n--- PAGE {(row["page_num"] + 1):03d} ---'
		current_page_number = row['page_num']
		current_line_number = 1
		current_blok_number = 1
	# if row['block_num'] > current_blok_number:
	# 	doc_string += '\n'
	# 	current_blok_number = row['block_num']
	# 	current_line_number= row['line_num']
	if abs(row['top'] + row['height'] / 2 - current_line_number) > 10:
		doc_string += '\n'
		current_line_number = row['top'] + row['height'] / 2
	
	doc_string += row['text'] + ' '

print(doc_string)
```
