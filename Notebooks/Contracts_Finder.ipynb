```python
import os
import pandas as pd
import csv

def find_sequence_with_pandas(file_path):

    df = pd.read_csv(file_path, header=None)
    y = df.loc[1,:].values[0]
    t = df.loc[1,:].values[1]
    

    no_contracts_seq = ["no", "contracts", "were", "entered"]
    no_contracts_seq2 = ["there", "were", "no", "city", "contracts"]
    # there were no City contracts relating to the Project Area. 
    no_contracts_index = 0
    no_contracts_index2 = 0
    
    yes_contracts_seq = ["description", "of", "certain", "external"]
    yes_contracts_seq2 = ["description", "of", "city", "contracts", "related"]
    yes_contracts_index = 0
    yes_contracts_index2 = 0
    
#__________________ no_contracts_seq 1 _____________________

    for index, row in df.iterrows():

        if row[12].lower() == no_contracts_seq[no_contracts_index]:
            no_contracts_index += 1  

            if no_contracts_index == len(no_contracts_seq):

                year = row[0]
                tif_district = row[1]
                return f'{year}*{tif_district}*-1'

                break  
        else:

            no_contracts_index = 0
            
 #__________________ no_contracts_seq 2 _____________________           
            

        if row[12].lower() == no_contracts_seq2[no_contracts_index2]:
            no_contracts_index2 += 1  

            if no_contracts_index2 == len(no_contracts_seq2):

                year = row[0]
                tif_district = row[1]
                return f'{year}*{tif_district}*-1'

                break  
        else:

            no_contracts_index2 = 0     
            
        
#__________________ yes_contracts_seq 1 _____________________        
        
        if row[12].lower() == yes_contracts_seq[yes_contracts_index]:
            yes_contracts_index += 1  

            if yes_contracts_index == len(yes_contracts_seq):

                year = row[0]
                tif_district = row[1]
                return f'{year}*{tif_district}*{row[2]}'

                break  
        else:

            yes_contracts_index = 0
        
        
#__________________ yes_contracts_seq 1 _____________________        
        
        
        if row[12].lower() == yes_contracts_seq2[yes_contracts_index2]:
            yes_contracts_index2 += 1  

            if yes_contracts_index2 == len(yes_contracts_seq2):

                year = row[0]
                tif_district = row[1]
                return f'{year}*{tif_district}*{row[2]}'

                break  
        else:

            yes_contracts_index2 = 0
            
#__________________ Neither Yes or no  _____________________

    return f'{y}*{t}*-2'





def parse_and_write_to_csv(folder_name, output_string):

    os.makedirs(folder_name, exist_ok=True)
    file_path = os.path.join(folder_name, "contracts.csv")
    file_exists = os.path.isfile(file_path)
    
    with open(file_path, mode='a', newline='') as file: 
        writer = csv.writer(file)
        if not file_exists:
            writer.writerow(['Year', 'TIF', 'Page'])
        if output_string: 
            data = output_string.split('*')
            writer.writerow(data)

def process_all_csv_files(input_folder, output_folder):
    for filename in os.listdir(input_folder):
        if filename.endswith(".csv"):
            file_path = os.path.join(input_folder, filename)
            contract_status = find_sequence_with_pandas(file_path)
            parse_and_write_to_csv(output_folder, contract_status)
            print(f"Processed {filename}")

input_folder = "csv_files"
output_folder = "contracts"


process_all_csv_files(input_folder, output_folder)
```

```python
# Initialize an empty dictionary to store the results
tif_dict = {}
n_dict = {}
x_dict = {}


for index, row in cont_df.iterrows():
    if row["Page"] > 0:
        if row["TIF"] in tif_dict:
            tif_dict[row["TIF"]].append(row["Year"])
        else:
            tif_dict[row["TIF"]] = [row["Year"]]
            

for index, row in cont_df.iterrows():
    if row["Page"] == -2:
        if row["TIF"] in n_dict:
            n_dict[row["TIF"]].append(row["Year"])
        else:
            n_dict[row["TIF"]] = [row["Year"]]
            
            
for index, row in cont_df.iterrows():
    if row["Page"] == -1:
        if row["TIF"] in x_dict:
            x_dict[row["TIF"]].append(row["Year"])
        else:
            x_dict[row["TIF"]] = [row["Year"]]

            
sorted_dict = {k: tif_dict[k] for k in sorted(tif_dict, key=lambda x: int(x))}
sorted_ndict = {k: n_dict[k] for k in sorted(n_dict, key=lambda x: int(x))}
sorted_xdict = {k: x_dict[k] for k in sorted(x_dict, key=lambda x: int(x))}
```
